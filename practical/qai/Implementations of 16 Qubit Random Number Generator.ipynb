```python
import random
from qiskit import QuantumCircuit, transpile, assemble, Aer

# Number of qubits
num_qubits = 16

# Create a quantum circuit with the specified number of qubits
qc = QuantumCircuit(num_qubits)

# Apply a Hadamard gate to each qubit to create superposition
for qubit in range(num_qubits):
    qc.h(qubit)

# Measure all the qubits
qc.measure_all()

# Compile the circuit for a quantum simulator
simulator = Aer.get_backend('qasm_simulator')
compiled_circuit = transpile(qc, simulator)

# Number of shots (measurement repetitions)
num_shots = 1024

# Run the circuit and get the results
job = simulator.run(compiled_circuit, shots=num_shots)
result = job.result()

# Extract and print the random numbers
counts = result.get_counts()
for binary, count in counts.items():
    decimal = int(binary, 2)
    print(f"Decimal: {decimal}, Binary: {binary}, Count: {count}")
```
